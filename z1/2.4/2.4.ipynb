```python
import numpy as np
import pylab as plt
```

```python
plt.rc('font', family = 'Times New Roman', size=10)
plt.rc('axes', unicode_minus = False)
plt.figure(figsize = (8, 5))
plt.subplot()

x=np.linspace(-10, 10, 100)
for k in range(1, 7):
    plt.subplot(2, 3, k)
    y = k*x**2+2*k
    plt.plot(x, y, '-b', label = 'k = '+str(k))
    plt.legend()
    plt.savefig('2.4.pdf')
```
