```python
import pandas as pd
import numpy as np
import pylab as plt
```

```python
df=pd.read_excel('附件1：区域高程数据.xlsx', sheet_name = 'Sheet1', header = None)
df
```

```python
Height_data = df[0:874]
Height_data
```

```python
z = Height_data.to_numpy().T
z
```

```python
x=np.arange(0, 43.651, 0.05)
y=np.arange(0, 58.201, 0.05)
X,Y = np.meshgrid(x, y)
```

```python
ax = plt.axes(projection = '3d')
ax.plot_surface(X, Y, z, cmap = 'coolwarm', alpha = 1)
ax.contour(X, Y, z, zdir = 'x', offset = 0, cmap = 'gist_earth')
ax.contour(X, Y, z, zdir = 'y', offset = 60, cmap = 'gist_earth')
ax.contour(X, Y, z, zdir = 'z', offset = 2200, cmap = 'gist_earth')
ax.set_xlabel('x', fontsize = 12)
ax.set_ylabel('y', fontsize = 12)
ax.set_zlabel('z', fontsize = 12)
plt.savefig('三维网格图.pdf')
```
