```python
import numpy as np
import pylab as plt
```

```python
plt.rc('font', family = 'Times New Roman', size=16)
plt.rc('axes', unicode_minus = False)
plt.figure(figsize = (6,5))
plt.subplot(111)

x=np.linspace(-10,10,100)
for k in range(1,7):
    y = k*x**2+2*k
    plt.plot(x, y, label = 'k='+str(k))
    plt.legend()

plt.savefig('2.3.pdf')
```
