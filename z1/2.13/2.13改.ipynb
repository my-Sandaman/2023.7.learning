```python
import pylab as plt
import numpy as np
```

```python
plt.rc('font', family = 'Times New Roman')
plt.rc('axes', unicode_minus = False)
plt.rc('font', size = 16)
plt.figure(figsize = (6, 5))
plt.subplot(111)
plt.rc('text', usetex = True)

t=np.linspace(0, 20, 1000)
x=np.linspace(0, 0, 1000)
e = 2.718281828459
U =-0.1*t*e**(-t)
v = x

plt.subplot(111)
plt.plot(t, U, '-c', label = '$U_{2}(t)=-0.1t\\mathrm{e}^{-t}$')
plt.plot(x, v, '-c')
plt.legend()
plt.savefig('2.13.pdf')
```
