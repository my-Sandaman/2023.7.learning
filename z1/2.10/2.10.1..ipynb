```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
import sympy as sp
```

```python
#绘制内侧曲线
# 创建数据点
x = np.linspace(-2, 2, 100)
y1 = np.linspace(1, 3, 100)
y2 = np.linspace(-1, 1, 100)
z1 = np.sqrt(4 * y1 - y1**2 - x**2)
z2 = np.sqrt(4 - x**2 - y2**2)

# 创建网格点
X, Y1 = np.meshgrid(x, y1)
X, Y2 = np.meshgrid(x, y2)

# 计算曲面
Z1 = np.sqrt(4 * Y1- Y1**2 - X**2)
Z2 = np.sqrt(4 - X**2 - Y2**2)

# 绘制一半合并曲面
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y1, Z1, cmap='viridis')
ax.plot_surface(X, Y2, Z2, cmap='magma')

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.savefig('内侧曲线.pdf')
```
