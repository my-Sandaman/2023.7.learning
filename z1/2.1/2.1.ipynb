```python
import pylab as plt
import numpy as np
```

```python
plt.rc('font', family = 'Times New Roman')
plt.rc('axes', unicode_minus = False)
plt.rc('font', size = 16)
plt.figure(figsize = (6,5))
plt.subplot(111)
plt.rc('text',usetex = True)

x1 = np.linspace(-2, 2, 1000)
e = 2.718281828459

#绘制chx
y1 = ((e**x1)+(e**-x1))/2
plt.plot(x1,y1,'-b',label='y=$\\frac{e^x+e^{-x}}{2}$')

#绘制shx
y2 = ((e**x1)-(e**-x1))/2
plt.plot(x1,y2,'-c',label='y=$\\frac{e^x-e^{-x}}{2}$')
#绘制0.5e^x

y3 = 1/2*(e**x1)
plt.plot(x1, y3, '-g', label = 'y=$\\frac{e^x}{2}$')
plt.legend()
plt.savefig('2.1.pdf')
```
