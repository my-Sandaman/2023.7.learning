```python
import numpy as np
from scipy.integrate import quad
```

```python
import matplotlib.pyplot as plt
```

```python

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
upper = np.inf
def fun46(t,x):
    return np.e**(-t)*t**(x-1)#定义积分函数fun46
x = np.arange(0.5,4.0,0.1)
I = [quad(fun46, 0, upper,args=i)[0] for i in x]#遍历每个x值,计算fun46的定积分,积分限为0到正无穷。quad返回一个tuple,取第0个元素就是积分值。这样通过循环可以获得每个x对应的积分值I
plt.plot(x, I, label='y=$\\int_0^{+\\infty}{\\mathrm{e}^{-t}t^{x-1}\\mathrm{d}x}$')
plt.legend(loc='upper left')#显示图例
plt.title('函数')
plt.xlabel('x')
plt.ylabel('I')
plt.show()
```
