```python
import numpy as np
import math
from scipy.integrate import quad
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

def fun46(t):
  return -0.1 * math.exp(-t) * t

t = np.arange(0.0, 10.0, 0.1)
I = [fun46(t) for t in t]

plt.plot(t, I, label='$U_2\\left( t \\right) $=$-0.1te^{-t}$')

ax = plt.gca()

# 设置右、下轴线不可见
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_color('none')

# 设置上、左轴线可见并调整位置
ax.spines['top'].set_position(('axes', 0.99))
ax.spines['left'].set_position(('axes', 0.01))

ax.set_xticklabels([]) # x轴刻度标签设置为空
ax.set_yticklabels([]) # y轴刻度标签设置为空
plt.xticks([])
plt.yticks([])#不设置单位尺度
ax.set_ylabel('$U_2\\left( t \\right) $', fontsize=14) # 设置y轴label
ax.set_xlabel('t', fontsize=14) # 设置x轴label

# 调整label位置
ax.xaxis.set_label_coords(1.0, 1.1)
ax.yaxis.set_label_coords(0, 0.05)


# 设置一个红色的'O'
ax.text(-0.1, 0, 'O', color='red')


plt.legend(loc='lower right')
plt.show()
```
