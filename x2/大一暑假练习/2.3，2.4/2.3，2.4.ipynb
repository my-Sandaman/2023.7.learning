```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
x = np.arange(-3, 3, 0.1)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)#plt.figure,fig.add_subplotax

for k in range(1,7):
    y = k*x**2 + 2*k
    ax.plot(x, y, label='k='+str(k))#k,y,plotlabel

ax.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Curves of y=$kx_2+2k$')

plt.show()
```

```python
fig = plt.figure()

for k in range(1,7):#range
    index = k
    ax = fig.add_subplot(2, 3, index)
    y = k*x**2 + 2*k
    ax.plot(x,y, label='k='+str(k))

    ax.legend()

plt.show()
```
