```python
table = df[0:874]
table
```

```python
#三维图
z = table.to_numpy().T
z
```

```python
x = np.arange(0,43.651,0.05)
y = np.arange(0,58.221,0.05)
```

```python
X,Y=np.meshgrid(x,y)#变成矩阵
```

```python
#设置画布
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,z,cmap='gist_earth',alpha=0.8)#颜色和透明度
ax.counter(X,Y,z,zdir='z',offset=2200,cmap=plt.cm.coolwarm)
ax.counter(X,Y,z,zdir='x',offset=0,cmap=plt.cm.coolwarm)
ax.counter(X,Y,z,zdir='y',offset=60,cmap=plt.cm.coolwarm)
```

```python
contr = plt.contour(x,y,z,cmap=plt.cm.gist_earth,linewidths=0.8)
plt.clabel(contr,fontsize=7.5,colors='k')#字体大小和颜色
plt.xlabel('x',fontsize=12)
plt.ylabel('y',fontsize=12)
xa = np.array([30,43])#传入两个点的横坐标
ya = np.array([0,30])#纵坐标
plt.plot(xa,ya,'p',color='r')#p是点的形状
plt.text(31,1,'A',fontsize=12)
plt.text(44,31,'B',fontsize=12)
plt.colorbar(shrink=0.8)
```

```python
# 生成一个高程矩阵
x = np.linspace(0, 50, 100)#使用np.linspace生成0-50之间,分成100个点的x、y坐标
y = np.linspace(0, 50, 100)
X, Y = np.meshgrid(x, y)#使用meshgrid将x,y坐标转换成网格矩阵X和Y
Z = np.hypot(X, Y)#计算每个网格点(X,Y)到原点的距离,作为高程矩阵Z

# 绘制contour图
plt.contour(X, Y, Z, 10)#绘制Z矩阵的等高线图,10表示分割的等高线数量

# 选择高程水平
h = 30

# 从contour图上测量该高程的轮廓线长度
len_h = 320 # 假设测量长度为320，选择高程水平h=30,测量该等高线长度len_h=320,相邻线间距d=5

# 测量相邻contour线间距
d = 5 # 假设为5

# 计算面积
S = len_h * d
print("Surface area is: ", S)

# 显示绘制的contour图
plt.savefig('contour.png')
plt.show()
```
