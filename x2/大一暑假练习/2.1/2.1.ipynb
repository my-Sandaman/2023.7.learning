```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
x = np.arange(-2,2, 0.01)
#- -5:数组的起始值,这里设置为 -5。 5 数组的结束值 不包含在数组中 这里设置为 5 。0.1 步长 即每两个连续元素之间的差值 这里设置为 0.1

# 函数
y1 = np.cosh(x)
y2 = np.sinh(x)
y3 = 0.5 * np.exp(x)

plt.plot(x, y1, label='y=cosh(x)')
#使用 x 数组作为横坐标,使用计算所得的 y1 数组作为纵坐标,绘制一条标签为"y=cosh(x)"的曲线。plot函数会自动将横纵坐标的值一一对应,从而绘制出函数曲线的图形。label参数用来在图例中显示该曲线的名称。
plt.plot(x, y2, label='y=sinh(x)')
plt.plot(x, y3, label='y=0.5e^x')

plt.legend()#显示图例
plt.title('三种函数')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
