```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')# 创建画布和坐标轴,创建一个画布并指定子图ax为3D坐标轴

x = np.linspace(-2, 2, 30)
y = np.linspace(-2, 2, 30)
X, Y = np.meshgrid(x, y)#用numpy的linspace生成x、y坐标向量,然后用meshgrid生成二维坐标网格

Z = np.sqrt(8 - 4*X**2 - 10*Y**2)#根据给定的方程,通过X、Y坐标计算对应Z坐标。


ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')#将坐标网格传给plot_surface绘制出曲面,用colormap上色

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')#给三个坐标轴添加标签

plt.savefig("X.pdf")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['font.sans-serif'] = ['Times New Roman']
plt.rcParams['axes.unicode_minus'] = False


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')# 创建画布和坐标轴,创建一个画布并指定子图ax为3D坐标轴

x = np.linspace(-2, 2, 30)
y = np.linspace(-2, 2, 30)
X, Y = np.meshgrid(x, y)#用numpy的linspace生成x、y坐标向量,然后用meshgrid生成二维坐标网格

Z = np.sqrt(X**2/4 + Y**2/6)#根据给定的方程,通过X、Y坐标计算对应Z坐标。


ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')#将坐标网格传给plot_surface绘制出曲面,用colormap上色

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')#给三个坐标轴添加标签
plt.tight_layout()
plt.savefig("Y.pdf")
```
